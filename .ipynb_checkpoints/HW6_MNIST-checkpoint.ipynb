```python
import numpy as np
import gzip
import struct


def load_images(filename):
    # Open and unzip the file of images:
    with gzip.open(filename, 'rb') as f:
        # Read the header information into a bunch of variables:
        _ignored, n_images, columns, rows = struct.unpack('>IIII', f.read(16))
        # Read all the pixels into a NumPy array:
        all_pixels = np.frombuffer(f.read(), dtype=np.uint8)
        # Reshape the pixels into a matrix where each line is an image:
        return all_pixels.reshape(n_images, columns * rows)
    # Prepare standardized MNIST inputs.
def standardize(training_set, test_set):
    average = np.average(training_set)
    standard_deviation = np.std(training_set)
    training_set_standardized = (training_set - average) / standard_deviation
    test_set_standardized = (test_set - average) / standard_deviation
    return (training_set_standardized, test_set_standardized)

# 60000 images, each 784 elements (28 * 28 pixels)
X_train_raw = load_images("mnist/train-images-idx3-ubyte.gz")

# 10000 images, each 784 elements, with the same structure as X_train
X_test_raw = load_images("mnist/t10k-images-idx3-ubyte.gz")

X_train = X_train_raw.reshape(-1,28,28)
X_test  = X_test_raw.reshape(-1,28,28)

X_train = np.divide(X_train, 255)
X_test = np.divide(X_test, 255)

print(X_train.shape)
print(X_test.shape)
```

```python
def load_labels(filename):
    # Open and unzip the file of images:
    with gzip.open(filename, 'rb') as f:
        # Skip the header bytes:
        f.read(8)
        # Read all the labels into a list:
        all_labels = f.read()
        # Reshape the list of labels into a one-column matrix:
        return np.frombuffer(all_labels, dtype=np.uint8).reshape(-1, 1)

def one_hot_encode(Y):
    n_labels = Y.shape[0]
    n_classes = 10
    encoded_Y = np.zeros((n_labels, n_classes))
    for i in range(n_labels):
        label = Y[i]
        encoded_Y[i][label] = 1
    return encoded_Y


# 60K labels, each a single digit from 0 to 9
Y_train_unencoded = load_labels("mnist/train-labels-idx1-ubyte.gz")

# 60K labels, each consisting of 10 one-hot encoded elements
Y_train = one_hot_encode(Y_train_unencoded)

# 10000 labels, each a single digit from 0 to 9
Y_test = load_labels("mnist/t10k-labels-idx1-ubyte.gz")
```

## Model

```python
def Softmax(x):
    x  = np.subtract(x, np.max(x))        # prevent overflow
    ex = np.exp(x)
    
    return ex / np.sum(ex)

def ReLU(x):
    return np.maximum(0, x)

def prepend_bias(X):
    return np.insert(X, 0, 1, axis=1)

from scipy.signal import convolve2d 
    
def Pool(x):
    (numFilters,xrow, xcol) = x.shape
    y = np.zeros((numFilters, int(xrow/2), int(xcol/2)))
    
    for k in range(numFilters):
        filter = np.ones((2,2)) / (2*2)
        image  = convolve2d(x[k, :, :], filter, 'valid')
        
        y[k, :, :] = image[::2, ::2]

    return y
```

```python
def initialize_weights(numFilters, kernel_size, n_input_variables, n_hidden_nodes, n_classes):
    
    w1 = 1e-2 * np.random.randn(numFilters, kernel_size, kernel_size) # 20, 9, 9
    
    w2_rows = n_input_variables + 1
    w2 = np.random.randn(w2_rows, n_hidden_nodes) * np.sqrt(1 / w2_rows)
    w3_rows = n_hidden_nodes + 1
    w3 = np.random.randn(w3_rows, n_hidden_nodes) * np.sqrt(1 / w3_rows)
    w4_rows = n_hidden_nodes + 1
    w4 = np.random.randn(w4_rows, n_hidden_nodes) * np.sqrt(1 / w4_rows)
    w5_rows = n_hidden_nodes + 1
    w5 = np.random.randn(w5_rows, n_classes) * np.sqrt(1 / w5_rows)
    
    return (w1, w2, w3, w4, w5)
```

```python
def Conv_create_feature_maps(x, W):
    (numFilters, wrow, wcol) = W.shape
    (xrow, xcol)             = x.shape
    
    yrow = xrow - wrow + 1  # Output width = (xrow - wrow + 2*Padding )/Stride + 1 
    ycol = xcol - wcol + 1 
    
    y = np.zeros((numFilters, yrow, ycol))
    
    for k in range(numFilters):
        filter = W[k, :, :]
        filter = np.rot90(np.squeeze(filter), 2) # 逆時針選轉兩次
        y[k, :, :] = convolve2d(x, filter, 'valid')
    
    return y
```

```python
def forward(X, w1, w2, w3, w4, w5): # X is 28 x 28
    y1 = Conv_create_feature_maps(X, w1)             # 20 x 20 x 20
    y2 = ReLU(y1)                                    # 20 x 20 x 20
    y3 = Pool(y2)                                    # 20 x 10 x 10
    y3_flat = np.reshape(y3, (1, -1)) # 1將y3變成一個 row, -1 是自動算 columns, 所以大小是 1x 2000
    a = np.matmul(prepend_bias(y3_flat), w2)        # (1x2001) x (2001x 100) 
    h1 = ReLU(a)                                    # 1 x 100 
    b  = np.matmul(prepend_bias(h1), w3)             # (1 x 101) x (101 x 10)
    h2 = ReLU(b)                                    # 1 x 100 
    c  = np.matmul(prepend_bias(h2), w4)     
    h3 = ReLU(c)                                    # 1 x 100 
    d  = np.matmul(prepend_bias(h3), w5)     
    y_hat = Softmax(d)                               #  1 x 10
    return (y_hat, d, h3, c, h2, b, h1, a, y3_flat, y3, y2, y1)
```

```python
def back(X, Y, y_hat, d, h3, c, h2, b, h1, a, y3_flat, y3, y2, y1,w1,w2,w3,w4,w5):
    
    dw5 = np.matmul(prepend_bias(h3).T, y_hat-Y) #為何不用除以X.shape[0]！！# (51 x 60000) x (60000 x 60000)  = 51 x 60000
    dh3 = np.matmul(y_hat - Y, w5[1:].T)                    # (60000 x 60000)  x (60000 x 50) = 60000  x 50
    dc  = dh3 * (h3 > 0)                                    # (60000 x 50) * (60000 x 50) = 60000  x 50
    dw4 = np.matmul(prepend_bias(h2).T, dc)    # (51 x 60000) x (60000 x 50) = 51 x 50
    dh2 = np.matmul(dc, w4[1:].T)                           # (60000 x 50) x (50 x 50)= 60000  x 50
    db  = dh2 * (h2 > 0)                                    # (60000 x 50) * (60000 x 50) = 60000  x 50
    dw3 = np.matmul(prepend_bias(h1).T, db)      # (51 x 60000) x (60000 x 50) = 51 x 50
    dh1 = np.matmul(db, w3[1:].T)                           # (60000 x 50) x (50 x 50)= 60000  x 50
    da  = dh1 * (h1 > 0)                                    # (60000 x 50) x (50 x 50)= 60000  x 50
    dw2 = np.matmul(prepend_bias(y3_flat).T, da)      # (785 x 60000) x (60000  x 50)= 785 x 50    
    dy3 = np.matmul(da, w2[1:].T)                        # (60000  x 50) x (50 x 50)= 60000  x 50
    dy3 = np.reshape(dy3, y3.shape)                       #  20 x 10 x 10
    
    dy2 = np.zeros_like(y2)                               # 20 x 20 x 20
    w6  = np.ones_like(y2)/(2*2)                          # 20 x 20 x 20
    dw1 = np.zeros_like(w1)                 
    for c in range(20):
        dy2[c,:,:] = np.kron(dy3[c,:, :], np.ones((2, 2))) * w6[c,:, :] #20x20x20  
    dy1 = dy2 * (y2 > 0)                                                #20x20x20
    for c in range(20): 
        dw1[c,:,:] = convolve2d(X, np.rot90(dy1[c,:,:], 2), 'valid')           
    return (dw1, dw2, dw3, dw4, dw5) 
```

```python
def loss(Y, y_hat):
    return - np.sum(Y * np.log(y_hat))
```

```python
def train(X_train, Y_train, X_test, Y_test, numFilters, kernel_size, n_hidden_nodes, 
          epochs, batch_size, lr):
    n_samples = X_train.shape[0]
    n_classes = Y_train.shape[1]
    n_input_variables=int(numFilters*(X_train.shape[1]-kernel_size+1)*(X_train.shape[2]-kernel_size+1)/4)                                  
    w1,w2,w3,w4,w5 = initialize_weights(numFilters, kernel_size, n_input_variables, n_hidden_nodes, 
                                    n_classes)
        
    batch_list = np.arange(0, n_samples, batch_size)
    
    beta  = 0.95    
    momentum1 = np.zeros_like(w1)
    momentum2 = np.zeros_like(w2)
    momentum3 = np.zeros_like(w3)
    momentum4 = np.zeros_like(w4)
    momentum5 = np.zeros_like(w5)
    
    for epoch in range(epochs):
        for batch in range(len(batch_list)):
            dw1 = np.zeros_like(w1)
            dw2 = np.zeros_like(w2)
            dw3 = np.zeros_like(w3)
            dw4 = np.zeros_like(w4)
            dw5 = np.zeros_like(w5)
            begin = batch_list[batch]
            training_loss = 0
            match_count = 0
            for idx in range(begin, begin+batch_size):
                # Forward pass = inference     
                y_hat, d, h3, c, h2, b, h1, a, y3_flat, y3, y2, y1 = forward(X_train[idx, :, :], w1, w2, w3, w4, w5)
                training_loss += loss(Y_train[idx,:], y_hat)
                label = np.argmax(y_hat, axis=1)
                match_count += (label == np.argmax(Y_train[idx,:]))
                
                # Backpropagation
                delta_w1, delta_w2, delta_w3, delta_w4, delta_w5 = back(X_train[idx, :, :], Y_train[idx,:], y_hat, d, h3, c, h2, b, h1, a, y3_flat, y3, y2, y1,w1, w2, w3, w4, w5)
                        
                dw1 = dw1 + delta_w1
                dw2 = dw2 + delta_w2    
                dw3 = dw3 + delta_w3
                dw4 = dw4 + delta_w4
                dw5 = dw5 + delta_w5
            
            dw1 = dw1 / batch_size
            dw2 = dw2 / batch_size
            dw3 = dw3 / batch_size
            dw4 = dw4 / batch_size
            dw5 = dw5 / batch_size
            if batch % 10 == 0:
                print("%5d-%d > Loss: %.8f, Accuracy: %.2f%%" %(epoch, batch, 
                                                        training_loss, match_count*100/batch_size))    

            momentum1 =  lr*dw1 + beta*momentum1
            w1  = w1 - momentum1
            momentum2 =  lr*dw2 + beta*momentum2
            w2  = w2 - momentum2
            momentum3 =  lr*dw3 + beta*momentum3
            w3  = w3 - momentum3
            momentum4 =  lr*dw4 + beta*momentum4
            w4  = w4 - momentum4
            momentum5 =  lr*dw5 + beta*momentum5
            w5  = w5 - momentum5
        
    return w1, w2, w3, w4, w5
```

```python
w1, w2, w3, w4, w5 =train(X_train, Y_train, X_test, Y_test, numFilters = 20, kernel_size = 9,
                  n_hidden_nodes=100, epochs=6, batch_size=100, lr=0.01)
print("Done.")
```

```python
n_samples = Y_test.shape[0]
match_count = 0
for idx in range(n_samples):
    y_hat, d, h3, c, h2, b, h1, a, y3_flat, y3, y2, y1 = forward(X_test[idx, :, :], w1, w2, w3, w4, w5)
    label = np.argmax(y_hat, axis=1)
    match_count += (label == Y_test[idx])
    
print("Accuracy is ", match_count*100/n_samples)
```
